#### Simplified Roulette Game :
A simplified version of Roulette game where we have 37 outcomes and we place bets on each roll of the ball for a single spin of the Roulette wheel. The goal is to calculate the expected value and standard deviation of winnings based on the bets.

Rules are as follows:
    Place a bet on one of the numbers on the Roulette wheel, each number has an equal probability of 1/37
    if the ball lands on the predicted numeber then win = 35,

```python
import random as rnd
import math
import matplotlib.pyplot as plt

# number of simulations
num_simulations = 1000

# function to create random integers from (0 to 36)
def spin_wheel():
    return rnd.randint(0,36)

# the outcome of the bet is determined
def bet_outcome(result):
    if (result > 30):
        return 35
    else:
        return 0

spin_results = []
bet_results = []

spin_results = [spin_wheel() for _ in range(num_simulations)] # spin wheel to record the result of spin
bet_results = [bet_outcome(result) for result in spin_results] # for each spin record the bets outcome
expected_value_spin = sum(spin_results)/num_simulations


gamble_result = []
for _ in range(num_simulations):
    spin_result = spin_wheel()
    bet_result = bet_outcome(spin_result)
    spin_results.append(spin_result)
    bet_results.append(bet_result)

expected_value_spin = sum(spin_results)/num_simulations
print("expected_value_spin = ", expected_value_spin)
expected_value_bet = sum(bet_results)/num_simulations
print("expected_value_bet = ", expected_value_bet)

spin_std_deviation = math.sqrt(sum((x - expected_value_spin) ** 2 for x in spin_results)/(num_simulations))
bet_std_deviation = math.sqrt(sum((x - expected_value_bet) ** 2 for x in bet_results)/(num_simulations))

plt.subplot(1,2,1)
plt.hist(spin_results, bins = 36, alpha = .5, label = "Spin Results", color = 'b')
plt.xlabel("Number on Roulette Wheel")
plt.ylabel("Frequency")
plt.title("Frequency of Roulette Spin Results")
plt.grid(True)
plt.legend()



# Plotting the frequency of bet outcomes
plt.subplot(1,2,2)
plt.hist(bet_results, bins = 36, alpha = .5, label = "Bet Results", color = 'g')
plt.xlabel("Number on Roulette Wheel")
plt.ylabel("Frequency")
plt.title("Frequency of Roulette bet Results")
plt.grid(True)
plt.legend()

plt.tight_layout(pad=2.0, w_pad=1.0, h_pad=1.0)
plt.show()
```

##### Roller Coaster Ride:

On a school trip, you and 100000 of your school mates go to an amusement park. There you find a couple of roller coasters. However, there is a catch to ride each of the roller coasters. You need to be tall to ride the roller coaster

- Only students with height greater than 100 cm can ride the Amazing Roller Coaster
- Only student with height between 80 cm and 100 cm can ride Blazing Roller Coaster
- The Amazing Roller Coaster(A) costs 10 Euros
- The Blazing Roller Coaster(B) costs 5 Euros

Calculate the cost paid by the school if you have 100000 students whose height has a mean of 100 cm and a std_dev of 20 associated with it, decides to go on the trip. Plot a distribution of experiment.

```python
# Your code
import matplotlib.pyplot as plt
import random

# Parameters
num_simulations = 100000
mean = 100
std_dev = 20

# Simulate test scores
heights = [random.gauss(mean, std_dev) for _ in range(num_simulations)]


# Calculate payments based on rides taken
cost = []
for height in heights:
    if height < 80:
        cost.append(0)
    elif 80 <= height < 100:
        cost.append(5)
    else:
        cost.append(10)
        

# Calculate average on payments
total_cost = sum(cost)
average_cost = total_cost / num_simulations

# Print results
print(f'Total cost in {num_simulations} simulations: ${total_cost}')
print(f'Average payment per simulation: {average_cost}')

# Create the first subplot
plt.subplot(1, 2, 1)
plt.hist(heights, bins=1000, alpha=0.5, label="height", color='g')
plt.xlabel("Height of Students")
plt.ylabel("Number of Students")
plt.title("Height of Students")
plt.grid(True)
plt.legend()

# Create the second subplot
plt.subplot(1, 2, 2)
plt.hist(cost, bins=[0, 1, 5, 6, 9, 10], alpha=0.5, label="cost", color='g')
plt.xlabel("Amount paid")
plt.ylabel("Number of students")
plt.title("Cost paid by the school distribution")
plt.grid(True)
plt.legend()

# adjust spacing between subplots
plt.tight_layout(pad=2.0, w_pad=1.0, h_pad=1.0)

# Display the figure with both subplots
plt.show()
```

### Features and labels
The aim is to predict the gender of the person based on characteristics such as height, weight. So we call what we want to predict the "label" and the characteristics the "features". This will be covered more later in the module, but is a good enough understanding for now.

```python
X = data.iloc[:,1:3] # X is the features in our dataset
y = data.iloc[:,0]   # y is the Labels in our dataset
```

### Train/Test Split

We have a bunch of data from the csv file. What we need to do is decide what of that data we will use to train the algorithm, and what of that data will be used to see how good the result is. More on this later in the module, but you can't use the same data you train with to test the outcome, as it will naturally fit that very well, but you won't catch things like overfitting. 

There is some science to this, but for now we will randomly select one third for testing, and the rest for training.

```python
################################
# Your code here
from sklearn.model_selection import train_test_split
(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.33)
```

### Train the algorithm
We won't go into the details here, but we will use a Gaussian Naive Bayesian. Use GaussianNB from sklearn to train the model. The Gaussian is needed as we're dealing with sparse, continuous data

```python
from sklearn.naive_bayes import GaussianNB

################################
# Your code here
model = GaussianNB().fit(X_train, y_train)

y_pred = model.predict(X_test)
print(y_pred)
```

### Test the algorithm
Using the test data

```python
from sklearn.metrics import accuracy_score

################################
# Your code here
acc_score = accuracy_score(y_test, y_pred)
print(acc_score)
```

# Application of Reinforcement Learning
Reinforcement Learning is the branch of Artificial Intelligence that works by training agents to make decision based on trial-and-error learning. It does this through a system of reward functions where by it tries to maximize the rewards obtained. There are many applications of reinforcement learning like
1. Medicine
    * In the field of surgery, considering the scarcity of highly efficient surgeons and their availability, RL can reduce the dependency on this. RL can also be used to get greater precision and accuracy and dexterity during surgery, there by reducing the patient recovery time. They can learn from experts at their specific domains such as appendectomy, laminectomy etc. The diagnosis and treatments suggested can be relatively bias free as its not a human that would be suggesting them and its based on previous data and it can learn from the outputs of previous attempts. RL can be used to guide doctors due to its ability to understand complex data inclusive of text, images of scans. The challenges to accomplish this would be to determine a parameters of reward function and how adaptable it would be and the stigma associated with machines acquiring confidential information, leading to privacy laws coming into effect. Although supervised learning is applicable here, it would take enormous amount of time and effort to acquire correct and valid data.

2. Education
    * The trainings provided can be personalized to achieve a more targeted training grounds for each individual as per their capability to comprehend the subject, rather than the traditional one size fits all approach. This leads to better comprehension of the material taught, where by we can increase the student engagement and motivation. By interacting with a user reinforcement learning algorithms can provide personalized feedback and provide resources to the user to increase his understanding of the topic. This also lets the student focus on their own learning needs and progress at their own speed. This is implemented in a few parts of the world with the adaptive assessments(even though they are not using the reinforcement learning), this is a possible place to incorporate it. The challenge here is the lack of high-quality data. It can even be difficult for the teachers to understand how the reinforcement model reached the conclusion of its analysis and why its predicting a defined solution. The ethical question of student's data being recorded is another challenge faced in the domain.
   
3. Reducing Energy Expenditure
    * Besides the efficiency aspects robots could be used to monitor and perform tasks that are harmful to humans. Eg. A nuclear accident. An example of the former would be the cooling of the Google Data Centers that are done by AI agents, which helped them reduce 40% energy expenditure. This works by taking snapshots of data from the data centers and sending it to a neural network which determines/predicts the effect of modification of the energy on future consumptions and identifying action to provide minimal power and implementing it. Although these are monitored by humans for safety, the human intervention is very minimal. This can be implemented not only on a large scale but also on a small scale like a house where by we can reduce dependency on power from the electrical grid and move to a renewable resource such as solar. This can lead to localized power heat maps that would be applicable on certain areas/communities using which the government can spend and the communities can understand their power consumption.
